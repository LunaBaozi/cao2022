```python
from collections import defaultdict
import re
import numpy as np
```

```python
aa_order = "ACDEFGHIKLMNPQRSTVWY"


```

```python
# Put your desired mutations here
# format:
# seqpos AAs
# comments work

muts = """
1 ACDE
2 PQ
10 STVI
11Q
42Q
44SAV
46NDS # You can comment like this
# 47M you can also comment out entire mutations
54LAV
"""

# PUT YOUR SEQUENCE HERE

native = "ARIENALAEELSAEFAREEELLELHRLIENFLEEAKRVKLTKALQGNGDPRLELAA"
```

```python
seqpos = defaultdict( lambda : [], {})
for i in range(len(native)):
    seqpos[i+1].append(native[i])

for row in muts.split("\n"):
    row = re.sub("#.*", "", row)
    row = row.strip().replace(" ", "")
    if (len(row) == 0):
        continue
    match = re.match("([0-9]+)([A-Z]+)", row)
    num = int(match.group(1))
    letters = match.group(2)
    for letter in letters:
        seqpos[num].append(letter)



rows = []
for pos in seqpos:
    row = []
    row.append("%i"%pos)
    for letter in aa_order:
        if ( letter in seqpos[pos] ):
            row.append("*")
        else:
            row.append("0")
    row.append("0")
    rows.append(row)
rows = np.array(rows)
for i in range(rows.shape[-1]):
    if ( i == 1 ):
        print(",".join(["-"]*len(rows)))
        print(",".join(["1"]*len(rows)))
    print(",".join(rows[:,i]))
        
        
# input for SwiftLib CSV format
# https://web.archive.org/web/20190222121816/http://rosettadesign.med.unc.edu/SwiftLib/
```
